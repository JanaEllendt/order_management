```python
import sqlite3
import uuid
import random
from datetime import datetime, timedelta

# Product catalog
product_catalogue = [
    {'articlenumber': 1234, 'articlename': 'Zahnbuerste', 'price': 3.99, 'currency': 'EUR'},
    {'articlenumber': 1245, 'articlename': 'Glas', 'price': 10.99, 'currency': 'EUR'},
    {'articlenumber': 4564, 'articlename': 'Brötchen', 'price': 0.6, 'currency': 'EUR'},
    {'articlenumber': 6241, 'articlename': 'Tastatur', 'price': 210.5, 'currency': 'EUR'},
    {'articlenumber': 5423, 'articlename': 'Maus', 'price': 90.8, 'currency': 'EUR'},
    {'articlenumber': 6501, 'articlename': 'nvidia geforce rtx 4090', 'price': 2389.00, 'currency': 'EUR'}
]

# Customer data
customers = [
    {"customer_id": str(uuid.uuid4()), "first_name": "Max", "last_name": "Mustermann", "email": "max.mustermann@example.com"},
    {"customer_id": str(uuid.uuid4()), "first_name": "Erika", "last_name": "Musterfrau", "email": "erika.musterfrau@example.com"},
    {"customer_id": str(uuid.uuid4()), "first_name": "John", "last_name": "Doe", "email": "john.doe@example.com"},
    {"customer_id": str(uuid.uuid4()), "first_name": "Jane", "last_name": "Doe", "email": "jane.doe@example.com"},
    {"customer_id": str(uuid.uuid4()), "first_name": "Hans", "last_name": "Meier", "email": "hans.meier@example.com"},
    {"customer_id": str(uuid.uuid4()), "first_name": "Anna", "last_name": "Schmidt", "email": "anna.schmidt@example.com"},
    {"customer_id": str(uuid.uuid4()), "first_name": "Peter", "last_name": "Müller", "email": "peter.mueller@example.com"},
    {"customer_id": str(uuid.uuid4()), "first_name": "Laura", "last_name": "Schneider", "email": "laura.schneider@example.com"},
    {"customer_id": str(uuid.uuid4()), "first_name": "Michael", "last_name": "Fischer", "email": "michael.fischer@example.com"},
    {"customer_id": str(uuid.uuid4()), "first_name": "Claudia", "last_name": "Weber", "email": "claudia.weber@example.com"},
    {"customer_id": str(uuid.uuid4()), "first_name": "Stefan", "last_name": "Wagner", "email": "stefan.wagner@example.com"},
    {"customer_id": str(uuid.uuid4()), "first_name": "Marie", "last_name": "Becker", "email": "marie.becker@example.com"},
    {"customer_id": str(uuid.uuid4()), "first_name": "Sebastian", "last_name": "Hoffmann", "email": "sebastian.hoffmann@example.com"},
    {"customer_id": str(uuid.uuid4()), "first_name": "Katrin", "last_name": "Schulz", "email": "katrin.schulz@example.com"},
    {"customer_id": str(uuid.uuid4()), "first_name": "Alexander", "last_name": "Koch", "email": "alexander.koch@example.com"},
    {"customer_id": str(uuid.uuid4()), "first_name": "Nina", "last_name": "Bauer", "email": "nina.bauer@example.com"},
    {"customer_id": str(uuid.uuid4()), "first_name": "Oliver", "last_name": "Richter", "email": "oliver.richter@example.com"},
    {"customer_id": str(uuid.uuid4()), "first_name": "Simone", "last_name": "Klein", "email": "simone.klein@example.com"},
    {"customer_id": str(uuid.uuid4()), "first_name": "Lukas", "last_name": "Wolf", "email": "lukas.wolf@example.com"},
    {"customer_id": str(uuid.uuid4()), "first_name": "Julia", "last_name": "Neumann", "email": "julia.neumann@example.com"}
]

# Database connection
conn = sqlite3.connect('orders.db')
cursor = conn.cursor()

# Create tables
cursor.execute('''
CREATE TABLE IF NOT EXISTS product (
    articlenumber INTEGER PRIMARY KEY,
    articlename TEXT NOT NULL,
    price REAL NOT NULL CHECK (price == ROUND(price, 2)),
    currency TEXT NOT NULL
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS customer (
    customer_id TEXT PRIMARY KEY,
    first_name TEXT NOT NULL,
    last_name TEXT NOT NULL,
    email TEXT NOT NULL,
    UNIQUE (first_name, last_name, email)
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS order_header (
    order_id TEXT PRIMARY KEY,
    customer_id TEXT NOT NULL,
    manufacturer TEXT NOT NULL,
    manufactur_place TEXT NOT NULL,
    shopnumber INTEGER NOT NULL,
    date TEXT NOT NULL,
    FOREIGN KEY (customer_id) REFERENCES customer (customer_id)
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS order_item (
    id TEXT PRIMARY KEY,
    order_id TEXT NOT NULL,
    articlenumber INTEGER NOT NULL,
    articlename TEXT NOT NULL,
    quantity INTEGER NOT NULL,
    price REAL NOT NULL CHECK (price == ROUND(price, 2)),
    currency TEXT NOT NULL,
    order_date TEXT NOT NULL,
    FOREIGN KEY (order_id) REFERENCES order_header (order_id),
    FOREIGN KEY (articlenumber) REFERENCES product (articlenumber)
)
''')

# Insert customers into customer table
for customer in customers:
    cursor.execute('''
    INSERT INTO customer (customer_id, first_name, last_name, email)
    VALUES (?, ?, ?, ?)
    ''', (customer['customer_id'], customer['first_name'], customer['last_name'], customer['email']))

# Insert products into product table
for product in product_catalogue:
    cursor.execute('''
    INSERT OR IGNORE INTO product (articlenumber, articlename, price, currency)
    VALUES (?, ?, ROUND(?, 2), ?)
    ''', (product['articlenumber'], product['articlename'], product['price'], product['currency']))

# Generate orders and order items for each month from January to 12th August 2024
start_date = datetime(2024, 1, 1)
end_date = datetime(2024, 8, 12)

while start_date <= end_date:
    for _ in range(10):  # 10 orders per month
        order_id = str(uuid.uuid4())
        customer = random.choice(customers)
        order_date = start_date.strftime('%Y-%m-%d')
        
        cursor.execute('''
        INSERT INTO order_header (order_id, customer_id, manufacturer, manufactur_place, shopnumber, date)
        VALUES (?, ?, ?, ?, ?, ?)
        ''', (order_id, customer['customer_id'], "Fielmann Group AG", "Hamburg", random.randint(100000000, 999999999), order_date))
        
        # Generate 20 items per order
        for _ in range(20):
            product = random.choice(product_catalogue)
            quantity = random.randint(1, 20)
            total_price = round(product['price'] * quantity, 2)
            
            cursor.execute('''
            INSERT INTO order_item (id, order_id, articlenumber, articlename, quantity, price, currency, order_date)
            VALUES (?, ?, ?, ?, ?, ROUND(?, 2), ?, ?)
            ''', (str(uuid.uuid4()), order_id, product['articlenumber'], product['articlename'], quantity, total_price, product['currency'], order_date))
    
    # Move to the next month
    start_date += timedelta(days=32)  # Adding 32 days to ensure month shift
    start_date = start_date.replace(day=1)

# Commit the changes and close the connection
conn.commit()
conn.close()

print("Data has been successfully inserted into the database.")
```
